```python
#building base line model

x = df.drop('charges', axis=1)
y = df['charges']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred_lr = lr.predict(x_test)
```

```python
r2_score(y_test, y_pred_lr)
```

```python
#building advance model
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
```

```python
print("R2:",r2_score(y_test,y_pred_rf))
print("MAE:",mean_absolute_error(y_test,y_pred_rf))
print("MSE:",mean_squared_error(y_test,y_pred_rf))
```

```python
#hyperparameter tuning
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}

grid = GridSearchCV(rf, param_grid, cv=5, scoring='r2')
grid.fit(x_train, y_train)
```

```python
best_model = grid.best_estimator_
```

```python
y_final=best_model.predict(x_test)
r2_score(y_test,y_final)
```

### Business Recommendations

1. Gender-neutral insurance policies can be continued as gender does not significantly impact claims.
2. Region-based pricing should be introduced due to varying claim behavior across locations.
3. Higher premium loading should be applied to smokers.
4. Discounts can be offered to customers with healthy BMI values.
5. Age-based risk adjustment should be applied instead of rejecting older customers.
6. Premiums should be adjusted based on the number of dependents.
7. Machine learning-based pricing strategies should be adopted for accurate risk assessment.
